```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn import metrics
df = pd.read_csv(r"C:\Users\Sudhindra Prakash\Downloads\AQI_Data.csv")
dfc = df.dropna()
X = dfc[['T', 'TM', 'Tm', 'SLP', 'H', 'VV', 'V', 'VM']]
Y = dfc['PM 2.5']
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, random_state=42)
model = LinearRegression()
model.fit(X_train, Y_train)
Y_pred = model.predict(X_test)
mae = metrics.mean_absolute_error(Y_test, Y_pred)
mse = metrics.mean_squared_error(Y_test, Y_pred)
r2 = metrics.r2_score(Y_test, Y_pred)
print("Model Evaluation:")
print("Mean Absolute Error (MAE):", mae)
print("Mean Squared Error (MSE):", mse)
print("R-squared (R2):", r2)
plt.scatter(Y_test, Y_pred, color='blue', alpha=0.3)
plt.plot(Y_test, Y_test, color='red')
plt.xlabel("Actual PM 2.5")
plt.ylabel("Predicted PM 2.5")
plt.title("Actual vs Predicted PM 2.5")
plt.show()
```
